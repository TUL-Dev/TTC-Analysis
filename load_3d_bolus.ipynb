```python
from pathlib import Path

import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
def read_scan(path: Path) -> np.ndarray:
    img = nib.load(path, mmap=False)
    proxy = np.asarray(img.dataobj, dtype=np.uint8).T
    return proxy
```

```python
masked_path = Path('/home/das/Downloads/Bolus/m948/m948d20150724BolusFull4DMasked.nii')
full_path = Path('/home/das/Downloads/Bolus/m948/m948d20150724BolusFull4D.nii')
```

```python
mask = read_scan(masked_path)
full = read_scan(full_path)
```
